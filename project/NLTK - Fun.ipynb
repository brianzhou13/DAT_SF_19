## Tokenization

Tokenization segments a document into its atomic elements. In this case, we are interested in tokenizing to words. This can be done with NLTK's tokenize.regexp module


```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
#Code above will match any word characters until it reaches a non-word character, like space.
#NLTK provides a number of pre-constructed tokenizers like nltk.tokenize.simple
#For unique cases, it's better to use regex and iterate until doc is accurately tokenized.

tokens_h1 = []
#list of list of tokens. 

#We will need a for-loop for this to go through the tokenization process.

for value in range(len(text_pd)):    
    tokens_h1.append(tokenizer.tokenize(text_pd.h1[0])) #tokenize h1
```

# Stemming

Stemming words is a technique used to reduce topically similar words to their root. The Porter stemming algorithm is the most widely used method. 

http://tartarus.org/~martin/PorterStemmer/

```python
from nltk.stem.porter import PorterStemmer

p_stemmer = PorterStemmer()
```

```python
#stem token

for value in range(len(tokens_h1)):
    tokens_h1[value] = [p_stemmer.stem(i) for i in tokens_h1[value]]

#texts = [p_stemmer.stem(i) for i in stopped_tokens]

#tokens_h1
```

# Constructing a Document-Term Matrix

Text should be cleaned (tokenized / stopped / stemmed) lists of lists of our documents. 

To generate a LDA model, we need to understand how frequently each term occurs within each document. To do that, we need to construct a document-term matrix with a package called gensim

```python
import gensim
from gensim import corpora, models


dictionary = corpora.Dictionary(tokens_h1)
#traverses tokens_h1, assigning a unique integer id to each unique token, while also collecting word-counts


corpus = [dictionary.doc2bow(value) for value in tokens_h1]

#the doc2bow() function converts dictionary into a bag-of-words

#the result, corpus, is a list of vectors equal to the number of documents. each document is a series of tuples.

#next the uniqueID must be converted into a 'bag-of-words' model 
#https://en.wikipedia.org/wiki/Bag-of-words_model
```

## Applying the LDA model

corpus is a document-term matrix, and we are ready to generate the LDA model:

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = 1, id2word = dictionary, passes = 20)

#num_topics is the number of requested latent topics to be extracted from the training corpus.
#id2word is a mapping from word ids (integers) to words (strings).
#passes is the number of laps the model will go through the corpus. more passes, the more accurate


#It is used to determine the vocabulary size, as well as for debugging and topic printing.
#get_document_topics(xxx) return topic distribution for the given document bow, as a list of (topic_id, topic_proba)
#get_topic_terms: Returns a list of (word_id, probability) 2-tuples for the most probable words in topic
#print_topics (num_topics=10, num_words=10)
```

```python
print(ldamodel.print_topics(num_topics=1, num_words=3))

ldamodel.print_topics(num_topics=10, num_words=10)

#we can see that we need to 
```
